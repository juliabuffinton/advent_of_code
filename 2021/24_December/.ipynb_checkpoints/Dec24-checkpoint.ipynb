```python
import math
```

```python
def parse_input(puzzle_input):
    with open(puzzle_input) as f:
        puzzle_input = [x.split(" ") for x in f.read().splitlines()]
    return puzzle_input
```

```python
def alu(instruction,var_state,value):
    
#     print("Instruction:",instruction)
    operator = instruction[0]
    a = instruction[1]
    
    # If it's only taking one argument then it has to be input
    if len(instruction) < 3:
        var_state[a] = value
    else:
        b = instruction[2]
        
        try: 
            b = int(b)
        except:
            b = var_state[b]
        
        if operator == "add":
            var_state[a] += b
        elif operator == "mul":
            var_state[a] *= b
        elif operator == "div":
            var_state[a] = math.floor(var_state[a] / b)
        elif operator == "mod":
            var_state[a] = var_state[a] % b
        elif operator == "eql":
            var_state[a] = int(var_state[a] == b)
                        
    return var_state
```

```python
def get_digit(number, n):
    return number // 10**n % 10
```

```python
def execute_instructions_and_check_valid(instructions,model_num):
    
    variables = {'w':0,'x':0,'y':0,'z':0}
    i = 13 # always 14 digit num
    instr_line = 1
    for instruction in instructions:
#         if i < 0:
#             break
        variables = alu(instruction,variables,get_digit(model_num,i))
        

        if instruction[0] == "inp":
            instr_line = 1
            i -= 1
            print("******************")
            print()
#         print("instruction:",instruction)
        print(f"{instr_line} variables:",variables)
        instr_line += 1

#         print()
    print(variables)
    # Return whether we've found a valid input
    return variables["z"] == 0
```

```python
def find_highest_valid_model_num(puzzle_input):
        
    instructions = parse_input(puzzle_input)
    
    for i,model_num in enumerate(range(99999999999999,10000000000000,-1)):
        
        if i % 500 == 0:
            print(f"tested {i} numbers")
        
        if '0' in str(model_num):
            continue
        else:
            valid = execute_instructions_and_check_valid(instructions,model_num)
            if valid:
                return model_num
    
    # If we find no valid numbers?!
    return model_num
```
