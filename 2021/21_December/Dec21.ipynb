```python
def deterministic_die_roll(die_val):
        
    spaces_to_move = 0
    
    # Roll the die three times
    for i in range(3):
        die_val += 1
        spaces_to_move += die_val
        
        # 100-sided die
        if die_val == 100:
            die_val = 0
        
    return die_val,spaces_to_move
```

```python
def calc_new_space(start_space,spaces_to_move):
    
    new_space = start_space + spaces_to_move
    
    # only go up to 10
    # Is there a better way to do this?!
    if new_space % 10 == 0:
        return 10
    else:
        return new_space % 10
```

```python
def deterministic_die_game(p0_start,p1_start):
    
    scores = {0:0,1:0}
    spaces = {0:p0_start,1:p1_start}
    
    curr_turn = 0 # player 1 starts
    die_val = 0
    num_rolls = 0
    
    while (scores[0] < 1000) and (scores[1] < 1000):
        die_val,move_spaces = deterministic_die_roll(die_val)
        new_space = calc_new_space(spaces[curr_turn],move_spaces)
    
        spaces[curr_turn] = new_space
        scores[curr_turn] += new_space
                
        curr_turn = abs(curr_turn-1)
        num_rolls += 3

    return scores,num_rolls
```

```python
def calc_output(dirac_game):
    
    final_scores, num_rolls = dirac_game
    
    loser_score = min(final_scores.values())
    
    return loser_score * num_rolls
```

```python
def dirac_game(start_positions):
    
    p0_start, p1_start = start_positions
    
    dirac_game = deterministic_die_game(p0_start,p1_start)
        
    return calc_output(dirac_game)
```
